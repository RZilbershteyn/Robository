```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.set_index('PassengerId')
data = data.drop(columns='PassengerId')
```

```python
data.head()
```

```python
y = data['Survived']
y.head()
```

```python
x = data.drop(columns='Survived')
x.head()
```

```python
x.loc[x['Sex'] == 'male', 'Sex'] = 0
x.loc[x['Sex'] == 'female', 'Sex'] = 1
x.head()
```

```python
x1 = x[['Sex', 'Age', 'SibSp', 'Parch']]
x1.head()
```

```python
x1.info()
```

```python
x = data[['Survived', 'Sex', 'Age', 'SibSp', 'Parch']].dropna()
x.loc[x['Sex'] == 'male', 'Sex'] = 0
x.loc[x['Sex'] == 'female', 'Sex'] = 1

x, x_test = train_test_split(x, test_size=0.2)

x1 = x[['Sex', 'Age', 'SibSp', 'Parch']]
y = x['Survived']

x1_test = x_test[['Sex', 'Age', 'SibSp', 'Parch']]
y_test = x_test['Survived']
```

```python
y.info()
```

```python
clf = RandomForestClassifier(n_estimators=50, min_samples_leaf=4)
clf.fit(x1, y)
clf.score(x1_test, y_test)
```
